## Step 1

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
x = np.linspace(-3, 3, 100).reshape(-1, 1)
print(x.shape)
```

```python
rng = np.random.RandomState(42)
y = np.sin(4 * x) + x + rng.uniform(size=len(x)).reshape(-1, 1)
```

```python
plt.figure(figsize=(8, 4), dpi = 300)
plt.plot(x, y, 'ro')
plt.xlabel('X', size = 15)
plt.ylabel('Y', size = 15)
plt.grid()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
polynomial1 = PolynomialFeatures(degree = 1)
polynomial2 = PolynomialFeatures(degree = 4)
polynomial3 = PolynomialFeatures(degree = 9)
polynomial4 = PolynomialFeatures(degree = 14)
```

```python
x_poly1 = polynomial1.fit_transform(x)
x_poly2 = polynomial2.fit_transform(x)
x_poly3 = polynomial3.fit_transform(x)
x_poly4 = polynomial4.fit_transform(x)
```

```python
# print(x[:10])
# print(x_poly[:10])
```

```python
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model2 = LinearRegression()
model3 = LinearRegression()
model4 = LinearRegression()
```

```python
model1.fit(x_poly1, y)
model2.fit(x_poly2, y)
model3.fit(x_poly3, y)
model4.fit(x_poly4, y)
```

```python
y_pred1 = model1.predict(x_poly1)
y_pred2 = model2.predict(x_poly2)
y_pred3 = model3.predict(x_poly3)
y_pred4 = model4.predict(x_poly4)
```

### Now you Can see different degree:

```python
plt.figure(figsize=(14, 10), dpi =300)

plt.subplot(2, 2, 1)
plt.plot(x, y, 'r.')
plt.plot(x, y_pred1, 'b')
plt.title('Degree = 1')
plt.grid()

plt.subplot(2, 2, 2)
plt.plot(x, y, 'r.')
plt.plot(x, y_pred2, 'b')
plt.title('Degree = 4')
plt.grid()

plt.subplot(2, 2, 3)
plt.plot(x, y, 'r.')
plt.plot(x, y_pred3, 'b')
plt.title('Degree = 9')
plt.grid()

plt.subplot(2, 2, 4)
plt.plot(x, y, 'r.')
plt.plot(x, y_pred4, 'b')
plt.title('Degree = 14')
plt.grid()
```
